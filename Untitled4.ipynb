```python
!pip install ultralytics opencv-python matplotlib

# Import required libraries
import torch
import cv2
import matplotlib.pyplot as plt
from ultralytics import YOLO

# Check if GPU is available
device = 'cuda' if torch.cuda.is_available() else 'cpu'
print(f'Using device: {device}')

# Load a pretrained YOLOv8 model
model = YOLO('yolov8n.pt')  # Using a small model for fast training

# Train on a dataset (replace 'your_dataset' with actual dataset path)
model.train(data='coco128.yaml', epochs=10, imgsz=640)  # Modify dataset path and epochs as needed

# Run inference on an image
img_path = 'https://ultralytics.com/images/zidane.jpg'  # Example image
results = model(img_path)

# Display the detected objects
for result in results:
    result.show()
    result.save('detected.jpg')  # Save output image

# Run inference on a video
video_path = 'your_video.mp4'  # Replace with your video file path
cap = cv2.VideoCapture(video_path)

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break
    results = model(frame)  # Run YOLO on the frame
    annotated_frame = results[0].plot()  # Draw results on frame
    cv2.imshow('YOLOv8 Detection', annotated_frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
